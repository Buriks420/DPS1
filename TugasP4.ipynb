```python
print("{:^44}". format("PT. DINGIN DAMAI"))
print("{:^44}". format("PROGRAM HITUNG GAJI KARYAWAN"))

nama = input("Nama Karyawan: ")
jabatan = int(input("Golongan Jabatan:"))
pendidikan = (input("Pendidikan: "))
lembur = int(input("Jumlah jam kerja: "))
gaji = 300000
honorlembur = 3500
if jabatan == 1:
    GolonganJabatan = 5
elif jabatan == 2:
    GolonganJabatan = 10
elif jabatan == 3 :
    GolonganJabatan = 15 
else :
    print("Masukan Golongan Jabatan yang benar")
    GolonganJabatan = 0
TunjanganJabatan = gaji * (1 + GolonganJabatan / 100)

if pendidikan == "sma" or pendidikan == "SMA":
    TunjanganPendidikan = 2.5
elif pendidikan == "d1" or pendidikan == "D1":
    TunjanganPendidikan = 5
elif pendidikan == "d3" or pendidikan == "D3":
    TunjanganPendidikan = 20
elif pendidikan == "s1" or pendidikan == "S1":
    TunjanganPendidikan = 30 
else :
    print("Masukan tingkat Pendidikan yang benar")
    TunjanganPendidikan = 0
TunjanganPendidikan = gaji * (1 + TunjanganPendidikan / 100)

if lembur > 8 :
    lembur = ((lembur - 8)*honorlembur)
else:
    lembur = 0

TotalGaji = TunjanganJabatan + TunjanganPendidikan + lembur
print(f"Karyawan yang bernama {nama}")
print("Honor yang diterima") 
print(f"    Tunjangan Jabatan    Rp {TunjanganJabatan:.0f}")
print(f"    Tunjangan Pendidikan Rp {TunjanganPendidikan:.0f}")
print(f"    Honor Lembur         Rp {lembur}")
print("                         ------------ +")
print(f"    Total Gaji           Rp {TotalGaji:.0f}")
```
