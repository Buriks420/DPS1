```python
print("GEROBAK FRIED CHICKEN")
print("-----------------------")
print("Kode JenisPotong Harga")
print("-----------------------")
print("D     Dada      Rp.2500")
print("P     Paha      Rp.2000")
print("S     Sayap     Rp.1500")
print("-----------------------")

BanyakJenis = int(input("Mau beli berapa banyak?"))
TotalHarga = 0
Pembelian = []


for i in range(BanyakJenis):
    print("Jenis ke - " + str(i+1))
    KodePotong = input("Kode Potong [D/P/S] : ").upper()
    if KodePotong == "D":
        Harga = 2500
        print("Dada Harganya Rp.", Harga)
    elif KodePotong == "P":
        Harga = 2000
        print("Paha Harganya Rp.", Harga)
    elif KodePotong == "S":
        Harga = 1500
        print("Sayap Harganya Rp.", Harga)
    else:
        print("Kode potong ayam tidak valid")
        continue
    BanyakPotong = int(input("Berapa banyak potongan yang ingin dibeli?"))
    JumlahBayar = Harga * BanyakPotong
    TotalHarga = TotalHarga + JumlahBayar

    Pembelian.append({
        "no": i+1,
        "Jenis Potong": KodePotong,
        "Harga Satuan": Harga,
        "Banyak Beli": BanyakPotong,
        "Jumlah Harga": JumlahBayar
    })
              
Pajak = (10/100) * TotalHarga
TotalHargaP = TotalHarga + Pajak
print("\nGEROBAK FRIED CHICKEN")
print("------------------------------------------------------------------------------------")
print("No.         Jenis Potong        Harga Satuan          Banyak Beli       Jumlah Harga")
print("------------------------------------------------------------------------------------")
for item in Pembelian:
    print(f"{item['no']} \t\t {item['Jenis Potong']} \t\t Rp.{item['Harga Satuan']} \t\t {item['Banyak Beli']} \t\t Rp.{item['Jumlah Harga']}")

print("------------------------------------------------------------------------------------")
print(f"\t\t\t\t\t\t\t   Jumlah Bayar: Rp.{TotalHarga}")
print(f"\t\t\t\t\t\t\t   Pajak 10%   : Rp.{Pajak:.0f}")
print(f"\t\t\t\t\t\t\t   Total Bayar : Rp.{TotalHargaP:.0f} ")
```
