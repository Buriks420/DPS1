```python
# Operator Penugasan
a = 2
b = 8
c = a+b
print(c)

# Operator Logika
d = True
e = True
f = d and e 
print(f)

# Operator Bitwise
g = 5  
h = 3  
i = g & h 
print(i)

# Operator Identitas
j = [1, 3, 4]
k = j
l = [1, 2, 3]
print(j is l)  

# Operator Keanggotaan
m = {"Buku, Bulpen, Pensil"}
n = {"Penghapus"}
print(m is n)
```
